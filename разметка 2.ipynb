Исправление : исходный

```python
import pandas as pd
from collections import Counter  
semslot = {}
mistakes = {}

data = pd.read_excel(r"C:\Users\lizao\Downloads\разметка.xlsx")
data.предложение = data.предложение.ffill()
for index, row in data.iterrows():
    if pd.isnull(data.loc[index, 'исправление semslot']) != True:
        if data.loc[index, 'исправление semslot'] in semslot.keys():
            semslot[data.loc[index, 'исправление semslot']].append(data.loc[index, 'ошибка'])
        else:
            semslot[data.loc[index, 'исправление semslot']] = []
            semslot[data.loc[index, 'исправление semslot']].append(data.loc[index, 'ошибка'])

for k, v in semslot.items():
    for t in v:
        if t.split()[8] in mistakes.keys():
            if k in mistakes[t.split()[8]]:
                mistakes[t.split()[8]][k] += 1 
            else:
                mistakes[t.split()[8]][k] = 1   
        else:
            mistakes[t.split()[8]] = Counter({k : 1})

mistakes
```

количество ошибок в семслоте

```python
mistakes_c = Counter()
for k, v in semslot.items():
    for t in v:
        if t.split()[8] in mistakes_c.keys():
            mistakes_c[t.split()[8]] += 1
        else:
            mistakes_c[t.split()[8]] = 1

mistakes_c = {k: v for k, v in sorted(mistakes_c.items(), key=lambda item: item[1], reverse=True)}
mistakes_c
```
