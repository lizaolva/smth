```python
import spacy
```

```python
nlp = spacy.load("ru_core_news_sm") # Load English tokenizer, tagger, parser and NER
```

```python
text='Проанализировать 1 предложение на том языке, который вы выбрали.'
```

```python
doc=nlp(text)  ##https://spacy.io/api/doc
```
